```python
df = df.drop(columns=['Unnamed: 0', 'userName','isEdited','title'], axis=1)
df
```

```python
# Grabs the first 200 reviews from each month

df_Jan = df[pd.to_datetime(df['date']).dt.month == 1]
df_Feb = df[pd.to_datetime(df['date']).dt.month == 2]
df_Mar = df[pd.to_datetime(df['date']).dt.month == 3]
df_Apr = df[pd.to_datetime(df['date']).dt.month == 4]
df_May = df[pd.to_datetime(df['date']).dt.month == 5]
df_June = df[pd.to_datetime(df['date']).dt.month == 6]
df_Jul = df[pd.to_datetime(df['date']).dt.month == 7]
df_Aug = df[pd.to_datetime(df['date']).dt.month == 8]
df_Sept = df[pd.to_datetime(df['date']).dt.month == 9]
df_Oct = df[pd.to_datetime(df['date']).dt.month == 10]
df_Nov = df[pd.to_datetime(df['date']).dt.month == 11]
df_Dec = df[pd.to_datetime(df['date']).dt.month == 12]
```

```python
# Creates new dataframes for each month with first 400 records from previous step

df1_Jan = df_Jan.head(400)
df1_Feb = df_Feb.head(400)
df1_Mar = df_Mar.head(400)
df1_Apr = df_Apr.head(400)
df1_May = df_May.head(400)
df1_June = df_June.head(400)
df1_Jul = df_Jul.head(400)
df1_Aug = df_Aug.head(400)
df1_Sept = df_Sept.head(400)
df1_Oct = df_Oct.head(400)
df1_Nov = df_Nov.head(400)
df1_Dec = df_Dec.head(400)
```

```python
# Places the dataframes into an array to be combined in the next step


sample_frames = [df1_Jan, df1_Feb, df1_Mar, df1_Apr, df1_May, df1_June, df1_Jul, df1_Aug,
                 df1_Sept,df1_Oct,df1_Nov,df1_Dec]
```

```python
# Combines the new dataframes from the above step for analysis

result = pd.concat(sample_frames)
result
```

```python
# Checks data types of columns
df.dtypes
```

```python
# Adds polarity column | Credits to Stack Overflow
def polarity_calc(text):
    try:
        return TextBlob(text).polarity
    except:
        return None
result['polarity'] = result['review'].apply(polarity_calc)
```

```python
# Adds subjectivity column | Credits to Stack Overflow
def subjectivity_calc(text):
    try:
        return TextBlob(text).subjectivity
    except:
        return None

result['subjectivity'] = result['review'].apply(subjectivity_calc)
```

```python
result
```

```python
from nrclex import NRCLex
import nltk 
# nltk.download('punkt')

emo_nrclex=[]
for i in result['review'].values.tolist():
  text_object = NRCLex(i)
  emotion= text_object.affect_frequencies
  # print(emo_nrclex)
  keymax =max(emotion, key=emotion.get)
  emo_nrclex.append(keymax)
  #print(emo_nrclex)
```

```python
result['NRCLEX']= emo_nrclex 
display(result)
result['NRCLEX'].describe()
```

```python
# Export final dataframe
result.to_csv("new_data.csv")
```
